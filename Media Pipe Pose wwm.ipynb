```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End
    
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    
    if angle >180.0:
        angle = 360-angle
        
    return angle 
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def calculate_angle(a, b, c):
    a = np.array(a)  # 첫 번째 점 (엉덩이)
    b = np.array(b)  # 두 번째 점 (무릎)
    c = np.array(c)  # 세 번째 점 (발목)
    
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360.0 - angle
    
    return angle

# 비디오 캡처 시작
cap = cv2.VideoCapture(0)
counter = 0
stage = None
wobble_counter = 0  # 균형 흔들림 카운터 초기화
sit_to_stand_counter = 0  # 앉았다 일어나는 카운터 초기화
curl_counter = 0  # 팔 굽혀 펴기 카운터 초기화

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # 이미지를 RGB로 색상 변환
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # 감지 실행
        results = pose.process(image)
        
        # 이미지를 다시 BGR로 변환
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 랜드마크 추출
        try:
            landmarks = results.pose_landmarks.landmark
            
            # 왼쪽 다리 관절의 좌표 가져오기
            hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
            
            # 각도 계산
            angle = calculate_angle(hip, knee, ankle)
            
            # 각도 시각화
            cv2.putText(image, str(angle), tuple(np.multiply(knee, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
            # 앉았다 일어나는 카운터 로직
            if angle > 170:
                stage = "stand"
            if angle < 90 and stage == 'stand':
                stage = "sit"
                counter += 1
                print(f"Sit-to-Stand Reps: {counter}")
        
            # 균형 흔들림 감지 로직
            if angle < 150 or angle > 210:  # 흔들림 감지를 위한 임계값 조정
                wobble_counter += 1
                print(f"Wobbles: {wobble_counter}")
        
        except Exception as e:
            print(f"Error: {e}")  # 에러 발생 시 출력
        
        # 앉았다 일어나는 카운터 렌더링
        # 상태 박스 설정
        cv2.rectangle(image, (0, 0), (225, 73), (245, 117, 16), -1)
        
        # 반복 데이터
        cv2.putText(image, 'REPS', (15, 12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, str(counter), (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # 단계 데이터
        cv2.putText(image, 'STAGE', (65, 12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, (60, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # 감지 결과 렌더링
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))
        
        cv2.imshow('WWM_test', image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def calculate_angle(a, b, c):
    a = np.array(a)  # 첫 번째 점 (엉덩이)
    b = np.array(b)  # 두 번째 점 (무릎)
    c = np.array(c)  # 세 번째 점 (발목)
    
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    
    if angle > 180.0:
        angle = 360.0 - angle
    
    return angle

# 비디오 캡처 시작
cap = cv2.VideoCapture(0)
counter = 0
stage = None
wobble_counter = 0  # 적절한 초기값으로 초기화

with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        
        # 이미지를 RGB로 색상 변환
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        # 감지 실행
        results = pose.process(image)
        
        # 이미지를 다시 BGR로 변환
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # 랜드마크 추출
        try:
            landmarks = results.pose_landmarks.landmark
            
            # 왼쪽 다리 관절의 좌표 가져오기
            hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
            
            # 각도 계산
            angle = calculate_angle(hip, knee, ankle)
            
            # 각도 시각화
            cv2.putText(image, str(angle), tuple(np.multiply(knee, [640, 480]).astype(int)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
            
            # 균형 흔들림 감지 로직
            if angle < 150 or angle > 210:  # 흔들림 감지를 위한 임계값 조정
                stage = "wobble"
                wobble_counter += 1
                print(f"Wobbles: {wobble_counter}")
            else:
                stage = "stable"
        
        except Exception as e:
            print(f"Error: {e}")  # 에러 발생 시 출력
        
        # 흔들림 카운터 렌더링
        # 상태 박스 설정
        cv2.rectangle(image, (0, 0), (225, 73), (245, 117, 16), -1)
        
        # 흔들림 데이터
        cv2.putText(image, 'WOBBLES', (15, 12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, str(wobble_counter), (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # 단계 데이터
        cv2.putText(image, 'STAGE', (65, 12), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
        cv2.putText(image, stage, (60, 60), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 2, cv2.LINE_AA)
        
        # 감지 결과 렌더링
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                  mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                  mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2, circle_radius=2))
        
        cv2.imshow('Balance Test', image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()

```

```python
def calculate_balance_score(wobble_counter):
    """
    Wobble counter에 따라 균형 검사 점수를 계산합니다.
    흔들림이 적을수록 높은 점수를 반환합니다.
    """
    if wobble_counter >= 100:
        return 4  # 흔들림이 많을 때 낮은 점수
    elif wobble_counter >= 50:
        return 3  # 약간 흔들림 발생 시
    else:
        return 2  # 거의 흔들리지 않음


def calculate_sit_to_stand_score(sit_to_stand_counter):
    """
    Sit-to-stand 카운터에 따라 점수를 계산합니다.
    횟수가 높을수록 빠르게 수행했다는 것으로 가정하여 높은 점수를 반환합니다.
    """
    if sit_to_stand_counter >= 10:
        return 4
    elif 5 <= sit_to_stand_counter < 10:
        return 3
    elif 2 <= sit_to_stand_counter < 5:
        return 2
    else:
        return 1


def calculate_curl_score(curl_counter):
    """
    Curl 카운터에 따라 덤벨 컬 점수를 계산합니다.
    횟수가 많을수록 높은 점수를 반환합니다.
    """
    if curl_counter < 5:
        return 2
    elif 6 <= curl_counter < 10:
        return 3
    else:
        return 4


def determine_sppb_stage(total_score):
    """
    총 점수를 통해 SPPB 단계를 판별합니다.
    """
    if total_score >= 10:
        return "SPPB 1단계 (매우 정상)"
    elif 7 <= total_score < 10:
        return "SPPB 2단계 (정상)"
    elif 4 <= total_score < 7:
        return "SPPB 3단계 (약간 위험)"
    else:
        return "SPPB 4단계 (높은 위험)"


def calculate_sppb_score(wobble_counter, sit_to_stand_counter, curl_counter):
    """
    각 테스트 카운터를 받아 SPPB 총 점수와 단계를 계산합니다.
    """
    # 각 테스트의 점수 계산
    balance_score = calculate_balance_score(wobble_counter)
    sit_to_stand_score = calculate_sit_to_stand_score(sit_to_stand_counter)
    curl_score = calculate_curl_score(curl_counter)

    # 총점 계산
    total_score = balance_score + sit_to_stand_score + curl_score

    # SPPB 단계 판별
    sppb_stage = determine_sppb_stage(total_score)

    # 결과 출력
    print(f"균형 검사 점수: {balance_score}")
    print(f"의자 일어서기 점수: {sit_to_stand_score}")
    print(f"덤벨 컬 점수: {curl_score}")
    print(f"총 SPPB 점수: {total_score}")
    print(f"SPPB 단계: {sppb_stage}")

    return total_score, sppb_stage


# 테스트용 카운터 값 입력
#wobble_counter = 5  # 예시 균형 검사 흔들림 횟수
#sit_to_stand_counter = 15  # 예시 의자 일어서기 카운터
#curl_counter = 4  # 예시 덤벨 컬 카운터

# SPPB 점수 및 단계 계산
total_score, sppb_stage = calculate_sppb_score(wobble_counter, sit_to_stand_counter, curl_counter)
```
